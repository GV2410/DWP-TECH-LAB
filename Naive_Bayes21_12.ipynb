```python
from sklearn import datasets
import pandas as pd
import statsmodels.api as sm
#import gaussian Bayes model
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score # Changed prediction_score to precision_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
import warnings
warnings.filterwarnings('ignore')
import numpy as np
```

```python
# prompt: iris dataset

import pandas as pd
iris = datasets.load_iris()
X = pd.DataFrame(iris.data)
y = pd.DataFrame(iris.target)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test,y_pred))
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
print("ROC_AUC score:",roc_auc_score(y_test,model.predict_proba(X_test),multi_class='ovr'))
```

```python
# prompt: split the dataset use model 1 by selecting the one variable create gaussian classifier and train the model using the training sets then mmodel prediction

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, roc_auc_score

# Assuming 'iris' DataFrame is already loaded and preprocessed as in the provided code.

# Select only one variable (e.g., 'sepal length (cm)')
X = iris[['sepal length (cm)']]
y = iris['species']

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Create and train the Gaussian Classifier
model = GaussianNB()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("ROC_AUC score:", roc_auc_score(y_test, model.predict_proba(X_test), multi_class='ovr'))
```

```python
# prompt: lets take all 4 columns and get the accuracy

# Assuming 'iris' DataFrame is already loaded and preprocessed as in the provided code.

# Select all four variables
X = iris[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
y = iris['species']

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=21)

# Create and train the Gaussian Classifier
model = GaussianNB()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("ROC_AUC score:", roc_auc_score(y_test, model.predict_proba(X_test), multi_class='ovr'))
```
