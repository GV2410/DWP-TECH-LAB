```python
# prompt: complete case analysis with definition and example

import pandas as pd
import numpy as np

# Definition:
# Complete case analysis (CCA), also known as listwise deletion, is a method for handling missing data in a dataset.
# It involves removing any data points that have missing values in any of the variables of interest.
# The analysis is then conducted only on the remaining complete cases.

# Example:

# Sample DataFrame with missing values
data = {'col1': [1, 2, np.nan, 4, 5],
        'col2': [6, 7, 8, np.nan, 10],
        'col3': [11, 12, 13, 14, 15]}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Perform complete case analysis
df_cca = df.dropna()
print(df.isna().sum())

print("\nDataFrame after Complete Case Analysis:")
df_cca
# Explanation:
# The original DataFrame has missing values (NaN) in 'col1' and 'col2'.
# The dropna() function removes any row with at least one NaN value.
# The resulting df_cca contains only complete cases, excluding rows with missing data.
```

```python
# prompt: Handling missing numerical data
# a) mean or median imputation with definition and get  age and gender dataset and when to use mean and median imputation

import pandas as pd
import numpy as np

# Sample DataFrame with missing age and gender data
data = {'age': [25, 30, np.nan, 40, 35, np.nan, 28, 22, np.nan, 45],
        'gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', np.nan]}
df = pd.DataFrame(data)

# Mean imputation for age
mean_age = df['age'].mean()
df['age_mean_imputed'] = df['age'].fillna(mean_age)

# Median imputation for age
median_age = df['age'].median()
df['age_median_imputed'] = df['age'].fillna(median_age)


# Mode imputation for gender (categorical data)
mode_gender = df['gender'].mode()[0] # mode() returns a Series, we take the first element
df['gender_mode_imputed'] = df['gender'].fillna(mode_gender)


print("Original DataFrame:\n", df)
print("\nDataFrame with mean/median/mode imputation:\n", df[['age', 'age_mean_imputed', 'age_median_imputed', 'gender', 'gender_mode_imputed']])


# When to use mean vs. median imputation:
# Mean imputation is generally preferred when the data is normally distributed and does not have significant outliers.
# Median imputation is preferred when the data is skewed or has outliers, as the median is less sensitive to extreme values.

# Example illustrating the difference:
# Create a dataset with an outlier
data_outlier = {'age': [25, 30, 32, 35, 38, 150]}  # 150 is an outlier
df_outlier = pd.DataFrame(data_outlier)
mean_age_outlier = df_outlier['age'].mean()
median_age_outlier = df_outlier['age'].median()
print(f"\nExample with outlier:")
print(f"Mean age with outlier: {mean_age_outlier}")
print(f"Median age with outlier: {median_age_outlier}")
```

```python
# prompt: b) End of distribution imputation (min and max) with definition and example with the same age dataset we took before replace min and max value

import pandas as pd
import numpy as np

# Sample DataFrame with age data
data = {'age': [25, 30, 28, 40, 35, 60, 22, 45]}
df = pd.DataFrame(data)

# Calculate min and max age
min_age = df['age'].min()
max_age = df['age'].max()

# Create new columns with min-max imputation
df['age_min_imputed'] = df['age'].replace(max_age,min_age)
df['age_max_imputed'] = df['age'].replace(min_age,max_age)


print("Original DataFrame:\n", df)
print("\nDataFrame with min-max imputation:\n", df[['age', 'age_min_imputed', 'age_max_imputed']])
```

```python
# prompt: c) Arbitrary value imputation with example and definition and take the same dataset of age and gender

import pandas as pd
import numpy as np

# Definition:
# Arbitrary value imputation replaces missing values with a pre-defined constant value.
# This method is simple but can introduce bias if the arbitrary value is not representative of the data.

# Example:
# Sample DataFrame with missing age and gender data
data = {'age': [25, 30, np.nan, 40, 35, np.nan, 28, 22, np.nan, 45],
        'gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', np.nan]}
df = pd.DataFrame(data)

# Arbitrary value imputation for age (e.g., using -1)
df['age_arbitrary_imputed'] = df['age'].fillna(-1)


# Arbitrary value imputation for gender (e.g., using 'Unknown')
df['gender_arbitrary_imputed'] = df['gender'].fillna('Unknown')

print("Original DataFrame:\n", df)
print("\nDataFrame with arbitrary value imputation:\n", df[['age', 'age_arbitrary_imputed', 'gender', 'gender_arbitrary_imputed']])
```
